```python
#@title 2.Функция для скачивания данных с Яндекс диска

# Базовая функция для скачивания данных с Яндекс диска
def Download_images_from_yandex_disk(public_key:str, name_zip_file:str):
  Base_URL = r'https://cloud-api.yandex.net/v1/disk/public/resources/download?'
  URL = Base_URL + urllib.parse.urlencode(dict(public_key=public_key))

  response = requests.get(URL)
  if response.status_code == 200:
    # Теперь здесь действительно JSON с полем 'href'
    Finale_URL = response.json().get('href')
    with open(name_zip_file, 'wb') as f:
      f.write(requests.get(Finale_URL).content)
  # Распаковка
  with zipfile.ZipFile(name_zip_file, 'r') as zip_ref:
    zip_ref.extractall(name_zip_file[:-4])

```

```python
#@title 6. Напишем основные функции для мэппинга
# Подключим статистические инструменты для построения регрессий
from sklearn.linear_model import Ridge
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures




# Функция извлекает данные из split_data_json
def Create_train_datasets_for_bottom_and_top(base_path:str, split_data:dict):
  Global_dict = {}
  for regime_of_learning, list_paths in split_data.items():
    Global_dict[regime_of_learning] = {}

    for source in ['bottom', 'top']:
      records = []

      for path in list_paths:
        # Формирование пути
        json_path = os.path.join(base_path, path, f'coords_{source}.json')

        # Проверка: вдруг в какой-то папке нет нужной камеры
        if not os.path.exists(json_path):
          continue

        with open(json_path, 'r', encoding='utf-8') as f:
          data = json.load(f)

        for pair in data:
          door_pts = {pt['number']: pt for pt in pair.get('image1_coordinates', [])}
          cam_pts = {pt['number']: pt for pt in pair.get('image2_coordinates', [])}

          # Ищем общие номера точек
          common_numbers = set(door_pts.keys()).intersection(set(cam_pts.keys()))

          for number in common_numbers:
            # Формируем готовую строчку для датафрейма
            records.append({
                            f'x_{source}': cam_pts[number]['x'],
                            f'y_{source}': cam_pts[number]['y'],
                            'x_door2': door_pts[number]['x'],
                            'y_door2': door_pts[number]['y']
                        })
            Global_dict[regime_of_learning][source] = pd.DataFrame(records)
  return Global_dict



def calculate_med(Y_true, Y_pred):
    distances = np.linalg.norm(np.array(Y_true) - np.array(Y_pred), axis=1)
    return np.mean(distances)





# Теперь после того как мы сформировали датасеты, мы можем применять различные модели для оценки предсказаняи
class Models_for_mapping:
  # Инициализируем массив
  def __init__(self,Train_dataset:pd.DataFrame,Val_dataset:pd.DataFrame):
    self.Train_dataset=Train_dataset
    self.Val_dataset=Val_dataset

  # Теперь реализуем модель полимиальной регерссии для нашей задачи
  def Regression_model_bottom(self):
     # Обучает полиномиальную регрессию 3-й степени и считает метрику MED
     X_train=self.Train_dataset['bottom'][['x_bottom', 'y_bottom']].to_numpy()
     Y_train=self.Train_dataset['bottom'][['x_door2', 'y_door2']].to_numpy()
     self.model_bottom = make_pipeline(PolynomialFeatures(degree=3),StandardScaler(), Ridge(alpha=1.0))
     self.model_bottom.fit(X_train, Y_train)

     X_val=self.Val_dataset['bottom'][['x_bottom', 'y_bottom']].to_numpy()
     Y_val=self.Val_dataset['bottom'][['x_door2', 'y_door2']].to_numpy()
     # Предикт на валидации
     Y_pred = self.model_bottom.predict(X_val)

     # Расчет Mean Euclidean Distance (MED)
     distances = np.linalg.norm(Y_pred - Y_val, axis=1)
     self.med_bottom = np.mean(distances)

  def Regression_model_top(self):
     # Обучает полиномиальную регрессию 3-й степени и считает метрику MED
     X_train=self.Train_dataset['top'][['x_top', 'y_top']].to_numpy()
     Y_train=self.Train_dataset['top'][['x_door2', 'y_door2']].to_numpy()


     self.model_top = make_pipeline(PolynomialFeatures(degree=3),StandardScaler(), Ridge(alpha=1.0))
     self.model_top.fit(X_train, Y_train)

     X_val=self.Val_dataset['top'][['x_top', 'y_top']].to_numpy()
     Y_val=self.Val_dataset['top'][['x_door2', 'y_door2']].to_numpy()
     # Предикт на валидации
     Y_pred = self.model_top.predict(X_val)


     # Расчет Mean Euclidean Distance (MED)
     distances = np.linalg.norm(Y_pred - Y_val, axis=1)

     self.med_top = np.mean(distances)

  def Predict_bottom(self, x: float, y: float):
    Y_pred = self.model_top.predict([[x],[y]])
    return Y_pred

  def Predict_top(self, x: float, y: float):
    Y_pred = self.model_top.predict([[x],[y]])
    return Y_pred


```

```python
Models=Models_for_mapping(Dataset['train'],Dataset['val'])
Models.Regression_model_bottom()
Models.Regression_model_top()
```
